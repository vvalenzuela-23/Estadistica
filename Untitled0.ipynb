### Análisis de la distribución y anomalías de 'PESO (Gramos)' y 'TALLA (Centímetros)'

#### PESO (Gramos):
*   **Histograma**: El histograma de 'PESO (Gramos)' muestra una distribución que se aproxima a una forma de campana, con una ligera asimetría hacia la izquierda (sesgo negativo), lo que sugiere que hay más valores hacia el extremo superior del peso. La mayor concentración de nacimientos se encuentra en el rango de los 3000 a 4000 gramos, indicando esta como la tendencia central. La dispersión del peso parece razonable para datos de nacimientos.
*   **Diagrama de Caja**: El diagrama de caja para 'PESO (Gramos)' confirma la tendencia central en torno a los 3000-3500 gramos (mediana). Se observan varios puntos individuales por debajo del bigote inferior, lo que indica la presencia de valores atípicos (bebés con peso muy bajo). También se aprecian algunos valores atípicos por encima del bigote superior, aunque en menor medida, sugiriendo pesos inusualmente altos.

#### TALLA (Centímetros):
*   **Histograma**: El histograma de 'TALLA (Centímetros)' también presenta una distribución que se acerca a una forma de campana, con una ligera asimetría hacia la izquierda (sesgo negativo). La mayoría de los nacimientos tienen una talla entre 45 y 55 centímetros. La tendencia central se ubica alrededor de los 50 centímetros. La dispersión es consistente con lo esperado para las tallas al nacer.
*   **Diagrama de Caja**: El diagrama de caja de 'TALLA (Centímetros)' muestra una mediana cercana a los 50 centímetros. Al igual que con el peso, se identifican valores atípicos tanto por debajo como por encima de los bigotes, representando bebés con tallas inusualmente bajas o altas, respectivamente. Sin embargo, los valores atípicos extremos son menos numerosos en comparación con los del peso.

#### Resumen General:
Ambas distribuciones, 'PESO (Gramos)' y 'TALLA (Centímetros)', exhiben una forma unimodal con una tendencia central bien definida. Presentan una ligera asimetría negativa, indicando que hay una cola más larga de valores más bajos. Es importante destacar la presencia de valores atípicos en ambos casos, especialmente en el extremo inferior, lo que podría corresponder a nacimientos prematuros o con bajo peso/talla para la edad gestacional.
```python
from scipy.stats import kstest

print("Imported kstest from scipy.stats.")
```

```python
import numpy as np
from scipy.stats import norm

# For 'PESO (Gramos)'
mean_peso, std_peso = original_df['PESO (Gramos)'].mean(), original_df['PESO (Gramos)'].std()
stat_peso_ks, p_peso_ks = kstest(original_df['PESO (Gramos)'].dropna(), 'norm', args=(mean_peso, std_peso))
print(f"Kolmogorov-Smirnov Test for PESO (Gramos): Statistic={stat_peso_ks:.3f}, p-value={p_peso_ks:.3f}")

# For 'TALLA (Centímetros)'
mean_talla, std_talla = original_df['TALLA (Centímetros)'].mean(), original_df['TALLA (Centímetros)'].std()
stat_talla_ks, p_talla_ks = kstest(original_df['TALLA (Centímetros)'].dropna(), 'norm', args=(mean_talla, std_talla))
print(f"Kolmogorov-Smirnov Test for TALLA (Centímetros): Statistic={stat_talla_ks:.3f}, p-value={p_talla_ks:.3f}")
```

```python
from scipy.stats import anderson

print("Imported anderson from scipy.stats.")
```
